# Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2*np.random.randn(100, 1)+3
rooms[0:10]
```

```python
price = 265 + 6*rooms + abs(np.random.randn(100,1))
price[0:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel('# of rooms, 2024 Average')
plt.ylabel('2024 average home price, $1000s USD')
```

```python
x= rooms
y = price

linearReg = LinearRegression()
linearReg.fit(x, y)

print(linearReg.intercept_, linearReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.68

*Estimated Coefficients*
- LinReg.coef_ = [6.01] Estimated coefficients for the terms in the linear regression problem.

```python
print(linearReg.score(x, y))
```
